#  3.1 Створіть змінну sentence і присвойте їй значення ‘she sells sea shells by the sea shore’  та напишіть фрагмент програми для виведення на екран всіх слів які починаються з ‘sh’.

```python
sentence = 'she sells sea shells by the sea shore' # ввід стрічки тексту
```

```python
phrase = sentence.split(' ') # перетворення в список
```

```python
sorted([w for w in set(phrase) if w.startswith('sh')]) # сортуємо текст по почту слів на СШ 
```

Програма повинна здійснювати наступну послідовність дій: створення початкової стрічки; створення стрічки, у якій буде зберігатися результат; for цикл для обробки стрічки символ за символом і запису неголосних символів в результуючу стрічку.
